```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import classification_report
```

# prep

Do a supersimple ML model with 3-4 variables

```python
df=pd.read_csv(r"../data\Sleep_health_and_lifestyle_dataset.csv")
df
```

```python
df["Sleep Disorder"].value_counts()
```

```python
#only numeric
x=df.drop(columns=["Person ID", "Sleep Disorder", "Gender", "Occupation", "Blood Pressure", "BMI Category"])
y=~df["Sleep Disorder"].isna()
```

```python
x
```

```python
x = x[["Age", "Sleep Duration"]]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=0)
```

# Train

```python
models={
    "Linear Regression": LinearRegression(),
    "DecisionTreeRegressor":DecisionTreeRegressor(),
    "RandomForestRegressor":RandomForestRegressor(),
}
```

```python
for model_name,model in models.items():
    print(f'Training {model_name}....')

    
    model.fit(x_train,y_train)

    y_pred=model.predict(x_test)
    y_pred= np.round(y_pred)
    
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("\n")
```

```python
model.feature_importances_
```

```python
import joblib

# save it
joblib.dump(model, 'model.joblib')
```

```python
#load

# Cargar
model_saved = joblib.load('model.joblib')
```

```python
load_pred = model_saved.predict(pd.DataFrame([[60, 8.1]], columns=["Age", "Sleep Duration"]))
np.round(load_pred)
```
